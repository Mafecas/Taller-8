**1. Ejemplo**
Graficaremos los datos en una figura, puede tener mas de un eje **Axes**, para la grafica se especificara un area en donde se pueda especificar coordenadas X y Y. Para crear un grafico con ejes utilizamos **pyplot.subplots** y para dibujar algunos datos en los ejes utilizamos **Axes.plot**

ax.plot([2, 5, 1, 4], [3, 2, 2, 3])
El primer argumento [2, 5, 3, 4] representa los valores en el eje x. En este caso, los puntos se ubicarán en las posiciones x: 2, 5, 3 y 4 respectivamente.
El segundo argumento [3, 2, 2, 3] representa los valores en el eje y correspondientes a los valores de x anteriores. Estos puntos se ubicarán en las posiciones y: 3, 2, 2 y 3 respectivamente.

```python
fig, ax = plt.subplots()  # Crea una figura que tenga un solo eje.
ax.plot([2, 5, 1, 4], [3, 2, 2, 3])  # Trazar algunos datos en los ejes..
```

**4.2.Hacer funciones auxiliares**

Utilizar los mismos graficos pero con diferentes datos, utilizaremos:

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

luego lo utilizaremos para complementar


```python
data1, data2, data3, data4 = np.random.randn(4, 80)  # 4 conjuntes de datos aleatorios
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

7.3. Trazar fechas y cadenas

Matplotlib puede manejar el trazado de matrices de fechas y matrices de cadenas, así como números de punto flotante.

```python


# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 4), constrained_layout=True)

# Generar datos de fechas y aleatorios
dates = np.arange(np.datetime64('2024-01-01'), np.datetime64('2024-02-01'), np.timedelta64(1, 'h'))
data = np.cumsum(np.random.randn(len(dates)))

# Graficar los datos
ax.plot(dates, data)

# Configurar el formato de las fechas en el eje x
cdf = mpl.dates.ConciseDateFormatter(ax.xaxis.get_major_locator())
ax.xaxis.set_major_formatter(cdf)

# Mostrar el gráfico
plt.show()

```

7.4.Objetos de eje adicionales

Al trazar datos de diferente magnitud en ocaciones se puede requerir un nuevo eje, este se puede agregar con la funcion de **tnix** nuevos ejes con un eje x invisible y para y **tniy** se agrega un nuevo eje a la derecha.

Puedes agregar un eje **secondary_xaxiso secondary_yaxis** que tenga una escala diferente al eje principal para representar los datos en diferentes escalas o unidades.
Ejemplo:

```python
t = np.linspace(0, 2*np.pi, 100)
s = np.sin(t)

# Crear la figura y los ejes
fig, (ax1, ax3) = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)

# Gráfico 1
l1, = ax1.plot(t, s)
ax2 = ax1.twinx()
l2, = ax2.plot(t, range(len(t)), 'C1')
ax2.legend([l1, l2], ['Sine (left)', 'Straight (right)'])

# Gráfico 2
ax3.plot(t, s)
ax3.set_xlabel('Angle [rad]')
ax4 = ax3.secondary_xaxis('top', functions=(np.rad2deg, np.deg2rad))
ax4.set_xlabel('Angle [°]')

plt.show()
```

**8.Datos mapeados en color**

En ocaciones podemos tener 3D en un mapa de colores.Matplotlib tiene varios tipos de gráficos que hacen esto:


```python
# Generar datos de ejemplo
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
data1 = np.random.randn(100)
data2 = np.random.randn(100)
data3 = np.random.rand(100)

# Crear la figura y los ejes
fig, axs = plt.subplots(2, 2, figsize=(10, 8), constrained_layout=True)

# Gráfico 1: pcolormesh()
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

# Gráfico 2: contourf()
co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

# Gráfico 3: imshow() con LogNorm()
pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma', norm=mpl.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow() with LogNorm()')

# Gráfico 4: scatter()
pc = axs[1, 1].scatter(data1, data2, c=data3, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')

plt.show()
```

8.1. Mapas de colores

Todos son ejemplos que derivan de **ScalarMappable**, se puede establecer un mapeo lineal entre vmin y vmax en el mapa de colores especificado por cmap.

8.2. Normalizaciones

En ocaciones queremos una asignación no lineal de los datos al mapa de colores, como en el **LogNorm**

8.3.Barras de colores

Agregar a **colorbar** proporciona una clave para relacionar el color con los datos subyacentes.

**9.Trabajar con múltiples figuras y ejes**
Se pueden abrir variasa figuras utilizando cualquiera de las **figuras.fig = plt.figure()fig2, ax = plt.subplots()**
Se pueden agregar varios ejes **plt.subplots()**. Tambien se pueden desarrollar  deseños mas complejos con objetos Axes que abarquen columnas o filas, usando **subplot_mosaic.**

```python
fig, axd = plt.subplot_mosaic([['left', 'upright'],
                               ['lowleft', 'downright']])

axd['left'].set_title('left')
axd['upright'].set_title('upright')
axd['lowleft'].set_title('lowleft')
axd['downright'].set_title('downright')

plt.show()
```

Matplotlib tiene herramientas bastante sofisticadas para organizar ejes: consulte Organización de múltiples ejes en una figura y Composición de figuras compleja y semántica (subplot_mosaic)
